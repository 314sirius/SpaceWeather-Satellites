```python
columns=['Year', 'Day', 'Hour', 'Minute', 'Field_Mag_average, nT', 'Bz_GSE, nT', 'Speed, km/s', 'Proton_Density, n/cc','Flow_Pressure, nPa', 'S/C, Xgse,Re','S/C, Ygse,Re','S/C, Zgse,Re','AE-index, nT','SYM/H, nT','Proton_Flux_10', 'Proton_Flux_30', 'Proton_Flux_60']
len(columns)
```

```python
import pandas as pd
def load_space_data(file_path, columns, skiprows=0):
    """
    Loads a space weather data file (.lst) into a pandas DataFrame.

    Args:
        file_path (str): The path to the .lst file.
        columns (list): A list of column names for the DataFrame.
        skiprows (int): Number of rows to skip at the beginning of the file.

    Returns:
        pd.DataFrame: The loaded DataFrame.
    """
    df = pd.read_csv(file_path, sep=r'\s+', skiprows=skiprows, header=None)
    df.columns = columns
    return df

df2010_2012 = load_space_data(path+'omni_5min_2010_2012.lst', columns)
df2013_2017 = load_space_data(path+'omni_5min_2013_2017.lst', columns)
df2018_2021 = load_space_data(path+'omni_5min_2018_2021.lst', columns)
df2022_2026 = load_space_data(path+'omni_5min_2022_2026.lst', columns)
```

```python
df = pd.concat([df2010_2012, df2013_2017, df2018_2021, df2022_2026], ignore_index=True)
```

```python
df.shape
```

```python
import numpy as np

# Define the placeholder values to be replaced with NaN
placeholder_values = [99999.9, 99999.99, 9999.99, 999.99, 99.99]

# Replace placeholder values with NaN across the entire DataFrame
df.replace(placeholder_values, np.nan, inplace=True)

df.head()
```

```python
df.to_csv('SpaceWeather/SpaceWeather.csv', index=False)
```
